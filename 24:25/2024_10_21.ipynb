<h1>SVT</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
help(SVR)
```

```python
#Generate synthetic data
X = np.sort(2 * np.random.rand(55, 1), axis=0)
y = (X**2).ravel()

#Add some noise to the data
y = y + 0.5 * (0.5 - np.random.rand(55))

#Create an SVR model with a polynomial kernel
e = 0.2
svr = SVR(kernel='poly', degree=3, epsilon=e)

#Train the model on the data
svr.fit(X, y)

#Make the predictions on the data
y_pred = svr.predict(X)

#Plot the predicted values against the true values
plt.scatter(X[np.setdiff1d(np.arange(len(X)), svr.support_)], y[np.setdiff1d(np.arange(len(X)), svr.support_)], color='black', label='other data')
plt.scatter(X[svr.support_], y[svr.support_], color='green', label='support vectors')

plt.plot(X, y_pred, color='cornflowerblue', label='prediction')
plt.plot(X, y_pred + e, color='red', ls='dashed', label='$\epsilon$-tube')
plt.plot(X, y_pred - e, color='red', ls='dashed')

plt.legend()
```
